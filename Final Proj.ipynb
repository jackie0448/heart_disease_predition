```python

heart_data_new= heart_data.drop(['trestbps','chol','fbs','restecg','exang','thal' ], axis=1)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(heart_data_new, test_size=0.2, random_state=42) 
train.head
train_labels= train.iloc[:,-1]
train_data= train.drop(['target'], axis=1)
test_labels= test.iloc[:,-1]
test_data= test.drop(['target'], axis=1)
```

```python
#checking for null values
train_labels.isnull().sum()
```

```python
train_data.isnull().sum()
```

```python
test_labels.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
from sklearn.svm import SVC
clf = SVC(kernel="sigmoid", gamma='auto')
clf.fit(train_data,train_labels)
test_labels_predict_svc= clf.predict(test_data)
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve
print("precision for SVC :",precision_score(test_labels, test_labels_predict_svc))
print("recall for svc:",recall_score(test_labels, test_labels_predict_svc))
print("F1 for svc:",f1_score(test_labels, test_labels_predict_svc))
print('AUC for svc:',roc_auc_score(test_labels, test_labels_predict_svc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=100, max_leaf_nodes=4, n_jobs=-1) #-1 means skitlearn is gonna use as many processors you have to run
rnd_clf.fit(train_data, train_labels)
test_labels_predict_rnd_tree= rnd_clf.predict(test_data)
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve
print("precision for random tree :",precision_score(test_labels, test_labels_predict_rnd_tree))
print("recall for random tree :",recall_score(test_labels, test_labels_predict_rnd_tree))
print("F1 for random tree :",f1_score(test_labels, test_labels_predict_rnd_tree))
print('AUC for random tree:',roc_auc_score(test_labels, test_labels_predict_rnd_tree))
```

```python
#this shows random forest is a better model bc at its precision is higher. 
```
